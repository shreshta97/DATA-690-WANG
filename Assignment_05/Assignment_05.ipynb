## 11. Filter the smaller dataset further to only keep colleges from Maryland, Virgina, Washington DC, Delaware, and Pennsylvinia. Save the filtered dataset to a new file named "Five_States_Colleges.csv" for future analysis. Upload this file to your GitHub along with the notebook.

```
# Make a list of FIVE state Colleges including only Maryland, Virgina, Washington DC, Delaware and Pennsylvinia
Five_States_Colleges = small_df.loc[df['STABBR'] == ('MD')], small_df.loc[df['STABBR'] == ('VA')], small_df.loc[df['STABBR'] == ('DE')], small_df.loc[df['STABBR'] == ('PA')], small_df.loc[df['STABBR'] == ('DC')]
print (Five_States_Colleges)
```

```
# Save the 5 state list as a CSV file 
import csv

with open("Five_States_Colleges.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(Five_States_Colleges)
```
