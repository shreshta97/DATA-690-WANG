# Summary: 
1. Intro: basics of pandas data frames more indepth into iloc loc and ix 
2. Tidy data: most common problems in messy data sets
    - Column headers are values, not variable names.
    - Multiple variables are stored in one column.
    - Variables are stored in both rows and columns.
    - Multiple types of observational units are stored in the same table.
    - A single observational unit is stored in multiple tables.
3. Final thoughts: 
    - some of the techniques used for tidy data would be very useful in the final project we completed for this class. for instance the time series COVID data that I used required alot of manipulation but the melt function could have helped condense the code alot.
